```python
import sys
import imageio.v2 as iio
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib
matplotlib.use("nbagg")
```

```python
image = iio.imread('colored balls.png')
plt.imshow(image)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
r, g, b = np.dsplit(image,image.shape[-1])
ax.scatter(r,g,b, marker=(5,2), c=np.column_stack([r.flatten(), g.flatten(), b.flatten()])/255)

ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
```
